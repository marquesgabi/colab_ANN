```
carvao_new= carvao.iloc[:,[1,2,3,4,5]].copy() #iloc é um tipo de localizador e carvao_new[:, ] é uma matriz :=pega todas colunas e os numeros sao linhas para copiar
print(carvao_new.head())
```

```
print (carvao.max(axis=0)) #pega o maior valor de cada coluna (nesse caso o 0= significa que ele vai buscar em todas as linhas daquela coluna o maior valor)
```

```
print (carvao.min(axis=0)) #pega o menor valor de cada coluna
```

```
def normalizerows(M):    #def é para vc definir uma funcao
  num=len(M.columns) #vai contar quantas colunas tem em M
  for i in range(num):
    Max=M.iloc[:,[i]].max()
    Min=M.iloc[:,[i]].min()
    a=float((Max+Min))/2
    b=float((Max+Min))/2
    M.iloc[:,[i]]=M.iloc[:,[i]].apply(lambda x:(x-a)/b) #para cada linha e coluna ele vai aplicar a funcao lambda
  return M
```

```
carvao_new=normalizerows(carvao_novo)
```
